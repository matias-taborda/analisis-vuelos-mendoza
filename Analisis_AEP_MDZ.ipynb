```python
# ==========================================
# AN√ÅLISIS DE VUELOS AEROPARQUE ‚Üí MENDOZA
# Carga manual de vuelos
# Python puro - una sola celda
# ==========================================

# Estados √∫nicos (iguales para todas las aerol√≠neas)
HORARIO = "Horario"
ATRASO_LEVE = "Atraso leve"
ATRASO_GRAVE = "Atraso grave"
CANCELADO = "Cancelado"
A_CONSULTAR = "A consultar"

ESTADOS_VALIDOS = [
    HORARIO,
    ATRASO_LEVE,
    ATRASO_GRAVE,
    CANCELADO,
    A_CONSULTAR
]

# --------------------------
# Funciones auxiliares
# --------------------------

def convertir_a_minutos(hora):
    horas, minutos = hora.split(":")
    return int(horas) * 60 + int(minutos)

def clasificar_estado(minutos_atraso):
    if minutos_atraso is None:
        return A_CONSULTAR
    if minutos_atraso == "cancelado":
        return CANCELADO
    if minutos_atraso <= 30:
        return HORARIO
    elif minutos_atraso <= 60:
        return ATRASO_LEVE
    else:
        return ATRASO_GRAVE

# --------------------------
# Lista de vuelos
# --------------------------

vuelos = []

def agregar_vuelo(aerolinea, numero_vuelo, hora_programada, hora_real):
    if hora_real.lower() == "cancelado":
        estado = CANCELADO
        atraso = None
    elif hora_real.lower() == "consultar":
        estado = A_CONSULTAR
        atraso = None
    else:
        minutos_prog = convertir_a_minutos(hora_programada)
        minutos_real = convertir_a_minutos(hora_real)
        atraso = minutos_real - minutos_prog
        estado = clasificar_estado(atraso)

    vuelos.append({
        "aerolinea": aerolinea,
        "vuelo": numero_vuelo,
        "programada": hora_programada,
        "real": hora_real,
        "estado": estado
    })

# ==========================================
# CARGA MANUAL DE VUELOS (EJEMPLOS)
# ==========================================

# Aerol√≠neas Argentinas
agregar_vuelo("Aerol√≠neas Argentinas", "AR1402", "06:20", "07:10")
agregar_vuelo("Aerol√≠neas Argentinas", "AR1402", "10:55", "11:04")
agregar_vuelo("Aerol√≠neas Argentinas", "AR1414", "11:45", "13:01")
agregar_vuelo("Aerol√≠neas Argentinas", "AR1416", "13:55", "15:06")
agregar_vuelo("Aerol√≠neas Argentinas", "AR1422", "19:10", "19:21")

# JetSMART
agregar_vuelo("JetSMART", "WJ3070", "09:55", "10:08")
agregar_vuelo("JetSMART", "WJ3078", "15:59", "16:22")
agregar_vuelo("JetSMART", "WJ3082", "21:15", "21:19")

# Flybondi
agregar_vuelo("Flybondi", "FO5050", "06:20", "07:29")
agregar_vuelo("Flybondi", "FO5054", "13:40", "14:06")
agregar_vuelo("Flybondi", "FO5056", "19:00", "19:36")

# ==========================================
# AN√ÅLISIS DE DATOS
# ==========================================

estadisticas = {}

for vuelo in vuelos:
    aero = vuelo["aerolinea"]
    estado = vuelo["estado"]

    if aero not in estadisticas:
        estadisticas[aero] = {
            "total": 0,
            HORARIO: 0,
            ATRASO_LEVE: 0,
            ATRASO_GRAVE: 0,
            CANCELADO: 0,
            A_CONSULTAR: 0
        }

    estadisticas[aero]["total"] += 1
    estadisticas[aero][estado] += 1

# Mostrar resultados

for aero, datos in estadisticas.items():
    print(f"Aerol√≠nea: {aero}")
    print(f"  Total de vuelos: {datos['total']}")
    for estado in ESTADOS_VALIDOS:
        print(f"  {estado}: {datos[estado]}")
    print()

# ==========================================
# MEJOR Y PEOR AEROL√çNEA
# ==========================================

mejor = max(estadisticas, key=lambda a: estadisticas[a][HORARIO])
peor = max(
    estadisticas,
    key=lambda a: estadisticas[a][CANCELADO] + estadisticas[a][ATRASO_GRAVE]
)

print(f"Mejor aerol√≠nea (m√°s vuelos en horario): {mejor}")
print(f"Peor aerol√≠nea (m√°s cancelados y atrasos graves): {peor}")
```
